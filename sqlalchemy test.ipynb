```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_path = "data/final_msa_with_coordinates2.csv"
```

```python
ms_data = pd.read_csv(csv_path)
ms_data
```

```python
ms_data.rename(columns = {'Incident ID':'incident_id','Incident Date':'incident_date', 'State':'state', 'City Or County':'city_county',
                         'Address':'address', '# Killed':'number_killed', '# Injured':'number_injured', 'Operations':'operations'}, inplace = True)

ms_data
```

```python
ms_data = ms_data.drop(['operations','coordinates'], axis =1)
ms_data
```

```python
# ms_data['longitude'].dtypes
```

```python
ms_data = ms_data.drop_duplicates(subset = 'incident_id', keep = "first")
ms_data
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/mass_shootings"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
ms_data.to_sql(name='ms_data', con=engine, if_exists='append', index=False)
```
