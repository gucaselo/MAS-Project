```python
url_election = "https://en.wikipedia.org/wiki/2020_United_States_presidential_election"
election_df = pd.read_html(url_election)
election_df = election_df[19]
election_df.head()
```

```python
# # Close browser
# browser.quit()
```

```python
list(election_df.columns)
```

```python
election_df["election_win"] = ""

for index, row in election_df.iterrows():
    blue = row[('Biden/HarrisDemocratic', 'Votes')]
    red = row[('Trump/PenceRepublican', 'Votes')]
    
    if blue > red :
        election_df.loc[index, 'election_win'] = 'blue'
    elif red > blue:
        election_df.loc[index, 'election_win'] = 'red'
        
election_df
```

```python
list(election_df.columns)
```

```python
election_df = election_df.drop([('Biden/HarrisDemocratic', 'Votes'),
 ('Biden/HarrisDemocratic', '%'),
 ('Biden/HarrisDemocratic', 'EV'),
 ('Trump/PenceRepublican', 'Votes'),
 ('Trump/PenceRepublican', '%'),
 ('Trump/PenceRepublican', 'EV'),
 ('Jorgensen/CohenLibertarian', 'Votes'),
 ('Jorgensen/CohenLibertarian', '%'),
 ('Jorgensen/CohenLibertarian', 'EV'),
 ('Hawkins/WalkerGreen', 'Votes'),
 ('Hawkins/WalkerGreen', '%'),
 ('Hawkins/WalkerGreen', 'EV'),
 ('Others', 'Votes'),
 ('Others', '%'),
 ('Others', 'EV'),
 ('Margin', 'Votes'),
 ('Margin', '%'),
 ('Marginswing', '%'),
 ('Totalvotes', 'Totalvotes')], axis=1)

election_df
```
